```python
# read in data

%matplotlib notebook
from collections import OrderedDict as od
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from pickle import load, dump
    
tps_dic = load(open('tps_dic.save', 'rb'))

print("Word Cloud for dominant 5 topics in the news headlines:")

plt.figure(figsize = (10, 5))
plt.imshow(WordCloud().fit_words(tps_dic))
plt.axis("off")
plt.show()
```
